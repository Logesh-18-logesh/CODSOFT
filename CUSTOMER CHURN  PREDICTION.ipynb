```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import RobustScaler
```

```python
df = pd.read_csv(r"F:\Churn_Modelling.csv")
print(df.shape)
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Geography'] = df['Geography'].replace({'France':0,'Spain':1,'Germany':2})
df['Gender'] = df['Gender'].replace({'Male':0,'Female':1})
df = df.drop(['RowNumber','CustomerId','Surname'],axis = 1)
df.head(10)
```

```python
x = df.drop(['Exited'],axis = 1)
y = df['Exited']
print(f"X shape: {x.shape}")
print(f"Y shape: {y.shape}")
```

```python
rs = RobustScaler()
x_scaled = rs.fit_transform(x)
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,train_size = 0.7, test_size = 0.3 ,random_state=42)
print(f"X_train shape: {x_train.shape}")
print(f"X test shape: {x_test.shape}")
print(f"Y train shape: {y_train.shape}")
print(f"Y test shape: {y_test.shape}")
```

```python
rf = RandomForestClassifier(n_estimators = 40, random_state = 42)
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
acc_score = accuracy_score(y_test,y_pred)
print(f"Accuracy Score: {acc_score}")
```
