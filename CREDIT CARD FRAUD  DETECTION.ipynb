```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
train_data = pd.read_csv(r"F:\fraudTrain.csv")
test_data = pd.read_csv(r"F:\fraudTest.csv")
```

```python
train_data.head(10)
```

```python
train_data['trans_date_trans_time'] = pd.to_datetime(train_data['trans_date_trans_time'])
train_data['dob'] = pd.to_datetime(train_data['dob'])
test_data['trans_date_trans_time'] = pd.to_datetime(test_data['trans_date_trans_time'])
test_data['dob'] = pd.to_datetime(test_data['dob'])
```

```python
train_data['trans_date_trans_time_day'] = train_data['trans_date_trans_time'].dt.day
train_data['trans_date_trans_time_month'] = train_data['trans_date_trans_time'].dt.month
train_data['trans_date_trans_time_year'] = train_data['trans_date_trans_time'].dt.year
train_data['dob_year'] = train_data['dob'].dt.year

test_data['trans_date_trans_time_day'] = test_data['trans_date_trans_time'].dt.day
test_data['trans_date_trans_time_month'] = test_data['trans_date_trans_time'].dt.month
test_data['trans_date_trans_time_year'] = test_data['trans_date_trans_time'].dt.year
test_data['dob_year'] = test_data['dob'].dt.year
```

```python
train_data.drop(columns=['trans_date_trans_time', 'dob'], inplace=True)
test_data.drop(columns=['trans_date_trans_time', 'dob'], inplace=True)
```

```python
train_data['gender'] = train_data['gender'].map({'M': 0, 'F': 1})
test_data['gender'] = test_data['gender'].map({'M': 0, 'F': 1})
```

```python
scaler = StandardScaler()
train_data['amt'] = scaler.fit_transform(train_data['amt'].values.reshape(-1, 1))
test_data['amt'] = scaler.transform(test_data['amt'].values.reshape(-1, 1))
```

```python
drop_columns = ['Unnamed: 0', 'cc_num', 'merchant', 'first', 'last', 'street', 'city', 'state', 'zip', 'lat', 'long', 'city_pop', 'job', 'trans_num', 'unix_time', 'merch_lat', 'merch_long']
train_data.drop(columns=drop_columns, inplace=True)
test_data.drop(columns=drop_columns, inplace=True)
```

```python
X_train = train_data.drop('is_fraud', axis=1)
y_train = train_data['is_fraud']
X_test = test_data.drop('is_fraud', axis=1)
y_test = test_data['is_fraud']
```

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

```python
X_train, X_test = X_train.align(X_test, join='outer', axis=1, fill_value=0)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print("Random Forests Accuracy:", accuracy)
```
